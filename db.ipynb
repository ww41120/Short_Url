```python
import sqlite3

conn = sqlite3.connect('url.db')


def create_table():
    cursor = conn.cursor()
    table_sql = """
    create table url(
      id INTEGER PRIMARY KEY autoincrement NOT NULL ,
      short_url text NOT NULL,
      long_url text NOT NULL
    )
    """
    # table_sql is used to store short Url and long Url
    cursor.execute(table_sql)
    conn.commit()      

    table_sql = """
    create table url_index(
      id INTEGER PRIMARY KEY autoincrement NOT NULL,
      useless INTEGER
    )
    """
    #url_index table used to store the index to find the url
    cursor.execute(table_sql)
    conn.commit()     
    cursor.close()




def get_long_url(short_url):
  
   
    cursor = conn.cursor()
    sql = "select * from url where short_url='{short_url}'".format(short_url=short_url)
    cursor.execute(sql)
    rows = cursor.fetchall() 
    if len(rows) == 0:
        return None
    row = rows[0]
    info = tuple(row)
    return info[2]
#This function can use short URL to find the long URL

def get_max_index():
   
    cursor = conn.cursor()
    sql = "insert into url_index(useless)values(1)"
    result = cursor.execute(sql)
    index = result.lastrowid
    conn.commit()
    cursor.close()
    return index
#This function is to get the newest url


def add_url(long_url, short_url):
 
    cursor = conn.cursor()
    sql = "insert into url(short_url, long_url)values('{short_url}', '{long_url}')"
    sql = sql.format(short_url=short_url, long_url=long_url)
    cursor.execute(sql)
    conn.commit()
    cursor.close()
```
